```python
#Sex
newdf2['Sex'] = newdf2['Sex'].replace(['男性','女性'],['M','F'])

# ICUAdmission_Type
## 入室形式 ("予定" (Planned) = 0, "緊急" (Emergency) = 1, "ICUでの手技" (ICU Procedure) = 2)
newdf2['ICUAdmission_Type'] = newdf2['ICUAdmission_Type'].replace(['予定','緊急', 'ICUでの手技'],['0','1', '2'])


#ICUAdmission_Source
##入室経路 ("病棟" (Ward) = 0, "救急外来" (ER) = 1, "手術室" (OR) = 2, "CCU" = 3, "HCU" = 4, "NICU" = 5, "PICU" = 6, "他のICU" = 7, "転院直入" (Transfer) = 8)
#"病棟" = 0, "救急外来" = 1, "手術室" = 2, "CCU" = 3, "HCU" = 4, "NICU" = 5, "PICU" = 6, "他のICU" = 7, "転院直入" = 8)
newdf2['ICUAdmission_Source'] = newdf2['ICUAdmission_Source'].replace(['病棟','救急外来', '手術室', 'CCU','HCU', 'NICU', 'PICU','他のICU', '転院直入'],['0','1', '2', '3', '4', '5', '6', '7', '8'])


#ICUAdmission_Elective
##入室区分 ("非手術" = 0, (Non-Surgery), "予定手術" = 1, "緊急手術" = 0 (Emergency Surgery))
newdf2['ICUAdmission_Elective'] = newdf2['ICUAdmission_Elective'].replace(['非手術','予定手術', '緊急手術'],['0','1', '0'])

#Tracheostomy
##気管切開 ("No" = 0, "外科的" (surgical) = 1, "経皮的" (Percutaneous) = 1)
newdf2['Tracheostomy'] = newdf2['Tracheostomy'].replace(['No','外科的', '経皮的'],['0','1', '1'])

#Emerg_Call
##緊急コール, "No" = 0, "コードブルー" (Code Blue) = 1, "RRT/MET" (Rapid Response Team/Emergency Medical Team) = 1, "no" = 0)
newdf2['Emerg_Call'] = newdf2['Emerg_Call'].replace(['No','コードブルー', 'RRT/MET', 'no'],['0','1', '1', '0'])

#Hospital_Type
###病院のタイプ ("公的病院" (Non-Profit Hosp) =0, "公立大学" (Public Uni)=1, "公立病院" (Public Hospital) =2, "国立大学" (National Uni) =3, "国立病院" (National Hosp) =4, "私立大学" (Private uni)=5, "私立病院" (Private Hosp)=6)
newdf2['Hospital_Type'] = newdf2['Hospital_Type'].replace(['公的病院','公立大学', '公立病院', '国立大学', '国立病院', '私立大学', '私立病院'],['0', '1', '2', '3', '4', '5', '6'])

#Discharge Outcome (Mortality)
##"生存" = 0, "死亡" = 1, "転院" = 0)
newdf2['Discharge_Outcome'] = newdf2['Discharge_Outcome'].replace(['生存','死亡', '転院'],['0','1', '0'])

#ICUDischarge_Outcome  
##退室時転帰 ("病棟" = 0, "死亡" = 1, "退院" = 2, "CCU" = 3, "HCU" = 4, "NICU" = 5, "PICU" = 6, "他のICU" = 7, "転院" = 8)
newdf2['ICUDischarge_Outcome'] = newdf2['ICUDischarge_Outcome'].replace(['病棟','死亡', '退院', 'CCU','HCU', 'NICU', 'PICU','他のICU', '転院'],['0','1', '2', '3', '4', '5', '6', '7', '8'])

#DOB (=Dobutamine)
##"使用していない" (did not use) = 0, "使用した" (Used) = 1, "使用していた" (Used in the past) = 1)
newdf2['DOB'] = newdf2['DOB'].replace(['使用していない','使用した', '使用していた'],['0','1', '1'])

# DOA (=Dopamine)
#"使用していない" (did not use) = 0, "5microg/kg/min以下" (Less than 5microg/kg/min) = 1, "5microg/kg/minを超え15microg/kg/min以下" (Between 5 to 15 microg/kg/min) = 1), "15microg/kg/minを超える" (more than 15) = 1, "0 ＜ DOA ≦ 5" = 1, "5γ以下"=1, "0microg/kg/minを超え5microg/kg/min以下"(Less than 5)=1, "No"=0
newdf2['DOA'] = newdf2['DOA'].replace(['使用していない','5microg/kg/min以下', '5microg/kg/minを超え15microg/kg/min以下', '15microg/kg/minを超える', '0 ＜ DOA ≦ 5', '5γ以下', '0microg/kg/minを超え5microg/kg/min以下', 'No'],['0','1', '1', '1', '1', '1', '1', '0'])

#NOR (=Norepinephrine)
#"使用していない" (did not use) = 0, "0.1microg/kg/min以下" (Less than 0.1microg/kg/min) = 1, "0.1microg/kg/minを超える" (Greater than 0.1 microg/kg/min) = 1), "0.1γ以下" (less than 0.1) = 1, "0.15γを超える" (greater than 0.15) = 1, "0.2microg/kg/minを超える" (more than 0.2) =1, "0 = NOR" =0, "使用していな\nい"=0
newdf2['NOR'] = newdf2['NOR'].replace(['使用していない','0.1microg/kg/min以下', '0.1microg/kg/minを超える', '0.1γ以下', '0.15γを超える', '0.2microg/kg/minを超える', '0 = NOR', '使用していな\nい'],['0','1', '1', '1', '1', '1', '0', '0'])

#ADR (=Adrenaline)
#"使用していない" (did not use) = 0, "0.1microg/kg/min以下" (Less than 0.1microg/kg/min) = 1, "0.1microg/kg/minを超える" (Greater than 0.1 microg/kg/min) = 1), "使用した" (used) = 1, "0.16γを超える" (greater than 0.16) = 1, "使用していない\n0.1microg/kg/min以下" (Not used/ less than 0.1) =0, "0 = ADR" = 0, 使用していない" =0, "使用していな\nい"=0
newdf2['ADR'] = newdf2['ADR'].replace(['使用していない','0.1microg/kg/min以下', '0.1microg/kg/minを超える', '使用した', '0.16γを超える', '使用していない\n0.1microg/kg/min以下', '使用していない.', '使用していな\nい', '0 = ADR'],['0','1', '1', '1', '1', '0', '0', '0', '0'])


newdf2.head()

```
